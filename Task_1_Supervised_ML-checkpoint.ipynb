```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#Reading the data
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
data.head(5)
```

```python
#Check if there is any null dataset in the Dataset
data.isnull == True
```

```python
sns.set_style('whitegrid')
sns.scatterplot(y = data['Scores'], x= data['Hours'])
plt.title('Percentage of Marks vs Study Hours', size=20)
plt.ylabel('Percentage of Marks', size=15)
plt.xlabel('Study Hours', size=15)
plt.show()
```

# There seems to be a correlation between 'Marks Percentage' and ' Hours Studied'  so we will plot a regression line for confirmation

```python
sns.regplot(x= data['Hours'], y=data['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Percentage of Marks', size=15)
plt.xlabel('Study Hours', size=15)
plt.show()
print(data.corr())
```

# The variables are positively correlated

# Model Training

1. Data splitting


```python
#Lets define the x and y from the data
x= data.iloc[:, :-1].values
y= data.iloc[:, 1].values

#Data Splitting into two
train_x, val_x, train_y, val_y= train_test_split(x,y, random_state = 0)
```

2. Fitting the Data into the model

```python
regression = LinearRegression()
regression.fit(train_x, train_y)
print("-----Model Trained-----")
```

# Predicting the Percentage of Marks

```python
pred_y = regression.predict(val_x)
prediction = pd.DataFrame({'Hours':[i[0] for i in val_x], 'Predicted Marks' : [k for k in pred_y]})
prediction
```

# Comparing the Predicted Marks with the Actual Marks


```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks':pred_y})
compare_scores
```

# Comparing the Actual marks with the Predicted Marks

```python
plt.scatter(x=val_x, y=val_y, color='Red')
plt.plot(val_x, pred_y, color='Black')
plt.title('Actual vs Predicted', size= 25)
plt.ylabel('Percentage of Marks', size = 15)
plt.xlabel('Study Hours', size = 15)
plt.show()
```

# Model Evaluation

```python
#Calculating the model accuracy
print('Mean absolute error:',mean_absolute_error (val_y, pred_y))
```

When theres a small value of Mean absolute error it defines that the chances of error on the model are minimal

# Now we will predict the score of a student if that student studies for 9.25 hours per day

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}". format(round(answer[0],3)))
```

So here we can see that the regression model stated that if a student studies for 9.25 hours a day so that student is likely to score 93.89 marks.
